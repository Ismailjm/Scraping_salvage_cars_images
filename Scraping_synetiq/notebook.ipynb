```python
import time
import os
import requests
from selenium import webdriver # type: ignore
from selenium.webdriver.chrome.service import Service # type: ignore
from selenium.webdriver.common.by import By # type: ignore
import time
import pandas as pd
```

```python
driver = webdriver.Chrome()

url = 'https://auctions.synetiq.co.uk/auction/items/cars'
driver.get(url)
# Get the maximum number of pages

max_page = driver.find_element(By.XPATH, '/html/body/section[3]/div/div/div/div/div[3]/div/ul/li[last()-1]/a')
max_page = int(max_page.text)
car_posts = []
i = 1
```

```python
print(f"Total pages: {max_page}")
```

```python
car_data = []
```

```python
for page in range(1, max_page+1):
    url = f'https://auctions.synetiq.co.uk/auction/items/?x=0&type=1&tab=1&list=1&section_id=&make=&transmission=0&fuel=0&seller=&category=&location=0&sort=&time=0&distance=0&options=&layout=r20&search=&fp=0&page={page}'
    driver.get(url)
    car_items = driver.find_element(By.CSS_SELECTOR, 'div.row.no-gutters')
    car_titles  = car_items.find_elements(By.CLASS_NAME,'list_title')

    # collectiong the links of the cars
    for title in car_titles:
        car_link = title.find_element(By.TAG_NAME,'a')
        car_title = car_link.find_element(By.TAG_NAME,'h2').text
        car_href = car_link.get_attribute("href")
        car_data.append({'title': car_title, 'link': car_href})
        # print(car_href)
```

```python
print(f"Total cars: {len(car_posts)}")
```

```python
# Create a new directory for each make if it doesn't exist
if not os.path.exists("synetiq"):
    os.makedirs("synetiq")
```

```python
i = 3785
```

```python
# collecting images of each car
for car in car_posts[909:]:
    driver.get(car)
    time.sleep(1)
    img_dots = driver.find_element(By.CLASS_NAME, 'img_i')
    # Find all child elements of the parent element
    dots = img_dots.find_elements(By.XPATH,"./*")  # This will find all direct child elements

    # Get the number of child elements
    num_dots = len(dots)
    car_data.append({'num_images': num_dots})
    if num_dots >= 22:
        continue  # Skip this post if the number of dots is 22
    for dot in range(0, num_dots):
        dot_id = "dot_" + str(dot)
        dot_element = driver.find_element(By.ID, dot_id)
        dot_element.click()
        
        img = driver.find_element(By.ID, 'mainimg')
        img_src = img.get_attribute('src')
        image = img_src
        file_name =str(i)
        i += 1
        with open(os.path.join("synetiq", file_name + '.jpeg'), 'wb') as img_file:
            img_file.write(requests.get(image).content)
    print(car_posts.index(car) + 1, "of", len(car_posts), "done")
```

```python
df = pd.DataFrame(car_data)
df.to_csv('car_data.csv', index=False)
```
